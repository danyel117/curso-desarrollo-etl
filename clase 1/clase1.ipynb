```python
import pandas as pd

import pandasql as ps

    
def pysqldf(q):
    return ps.sqldf(q, globals())
```

```python
df = pd.read_csv('./world_bank_development_indicators.csv', sep=',')
```

```python
df
```

```python
# seleccion de una columna

df["country"]
```

```python
# seleccionar una lista de columnas

precipitacion_anual = df[['country', 'date', 'avg_precipitation']]


precipitacion_anual
```

```python
# seleccionar columnas por índice 

df.iloc[:,[0, 1, 2]]

# seleccionar los primeros 10 registros de las primeras 3 columnas

# acceder a un grupo de filas o columnas especificas: [0, 10, 5, 35]
# acceder a un rango de filas o columnas: indiceInicial : indiceFinal

df.iloc[ 0:10 , [25, 36, 2] ]

df.iloc[ [0,1,2] , 10:25 ]
```

```python
# acceder a ciertas columnas segun el nombre
df.loc[ 0:100, ['country','date','gini_index'] ]
```

```python
# seleccionar cierto tipo de dato

df.select_dtypes(include=['int', 'float'])
```

```python
df[['country','date']].rename(columns={'country':'país', 'date':'fecha'})
```

```python

pysqldf("""
select
country as pais,
date as fecha,
death_rate as tasa_mortalidad
from df
""")
```

```python
# filtrar por una columna

df[ df['country'] == 'Colombia' ] # comparadores logicos: ==, !=, <, <=, >, >=


df[ df['avg_precipitation'] > 3200 ]
```

```python
df[ (df['country'] == 'Colombia') & (df['avg_precipitation'] > 3200) ] # para hacer comparativos 'and' se usa el caracter &. Para hacer comparativos 'or' se usa el caracter |
```

```python
# filtro isin

df[ df['country'].isin(['Colombia', 'Ecuador']) ]
```

```python
# filtros con SQL

pysqldf("""
select * from df
where country = 'Colombia'
limit 10
""")
```

```python
# filtros con SQL

pysqldf("""
select * from df
where country = 'Colombia' and avg_precipitation >= 3240
limit 10
""")
```

```python
# filtro in

pysqldf("""
select * from df
where country in ('Colombia', 'Ecuador')
limit 10
""")
```

```python
# instrucciones de agrupacion

df[['country', 'gini_index']].groupby('country').max()
```

```python
pysqldf("""
select
country,
max(gini_index)
from df
group by country
""")
```

```python
pysqldf("""
select 
country,
avg(avg_precipitation)
from df
group by country
having avg(avg_precipitation) > 3000
""")
```
