```python
import pandas as pd

import pandasql as ps

import seaborn as sns
import matplotlib.pyplot as plt
    
from pandas_profiling import ProfileReport

def pysqldf(q):
    return ps.sqldf(q, globals())
    
```

```python
df = pd.read_csv('./world_bank_development_indicators.csv', sep=',')
```

```python
profile = ProfileReport(df, title='Exploración de indicadores Banco Mundial', explorative=True)

profile.to_file('./data_profile.html')
```

```python
# histogramas

dispersiones = pysqldf("""
                select 
                CO2_emisions,
                "renewvable_energy_consumption%",
                GDP_current_US
                from df
                where country = 'Colombia'
              """)

dispersiones.hist(bins=30, figsize=(20, 5))
```

```python
# histogramas

dispersiones = pysqldf("""
                select 
                life_expectancy_at_birth,
                death_rate
                from df
              """)

dispersiones.hist(bins=30, figsize=(20, 5))
```

```python
# grafico de dispersion

dispersion = pysqldf("""
                    select
                    life_expectancy_at_birth,
                    GDP_current_US
                    from df
                    where 
                    GDP_current_US  is not null and life_expectancy_at_birth is not null
                    and country in ('Colombia', 'Ecuador', 'Peru', 'Chile')
                    """)

sns.scatterplot(data = dispersion, x='GDP_current_US', y='life_expectancy_at_birth')
```

```python
# grafico de dispersion

dispersion = pysqldf("""
                    select
                    avg("agricultural_land%") as agricultural_land_avg,
                    avg(CO2_emisions) as co2_emissions_avg
                    from df
                    group by country
                    """)
#dispersion
sns.scatterplot(data = dispersion, x='agricultural_land_avg', y='co2_emissions_avg')
```

```python
# grafico de linea

evolucion_gdp_latam = pysqldf("""
            select
            country,
            strftime('%Y', date) as year,
            GDP_current_US
            from df
            where country in ('Colombia', 'Ecuador', 'Peru', 'Chile')
            """)

plt.figure(figsize=(12,6))
sns.lineplot(data=evolucion_gdp_latam, x='year', y='GDP_current_US', hue='country')
plt.title('GDP over time')
plt.xlabel('year')
plt.xticks(rotation=90)
plt.ylabel('GDP')
plt.show()
```

```python
# matriz de correlacion


df_para_correlaciones = pysqldf("""
        select 
        "agricultural_land%", "forest_land%", "land_area",
       "avg_precipitation", "trade_in_services%",
       "control_of_corruption_estimate", "control_of_corruption_std",
       "access_to_electricity%", "renewvable_energy_consumption%",
       "electric_power_consumption", "CO2_emisions",
       "other_greenhouse_emisions", "population_density", "inflation_annual%",
       "real_interest_rate", "risk_premium_on_lending"
        from df
        """)

matriz_de_correlacion = df_para_correlaciones.corr()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=matriz_de_correlacion, cmap='coolwarm', annot=True)
```
