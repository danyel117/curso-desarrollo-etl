```python
import pandas as pd

import pandasql as ps
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

def pysqldf(q):
    return ps.sqldf(q, globals())
    
```

```python
# cargar el dataset

df = pd.read_csv('./Spotify Most Streamed Songs.csv', sep=',')
```

```python
# Agrupar por el nombre del artista y contar cuántas canciones tienen en el dataset. Mostrar los 10 artistas con más canciones.

pysqldf("""
        select 
        "artist(s)_name",
        count(*) as count
        from df
        group by "artist(s)_name"
        order by count desc
        limit 10
        """)
```

```python
# Calcular el promedio de streams para canciones en modo mayor y menor.
pysqldf("""
        select 
        mode,
        avg(streams) as promedio
        from df
        group by mode
        """)
```

```python
# Crear un gráfico de dispersión para ver la relación entre "danceability" y el número de streams.

filtered_df = pysqldf("""
        select 
        "danceability_%",
        cast(streams as int) as streams
        from df
        where cast(streams as int) > 0
        """)

sns.scatterplot(data=filtered_df, x='danceability_%', y='streams')

```

```python
# Agrupar por la columna key y calcular el promedio de streams para cada key.
pysqldf("""
        select
        key,
        avg(streams)
        from df
        where cast(streams as int) > 0
        group by key
        """)
```

```python
# Crear un histograma que muestre cómo se distribuye la energía de las canciones.
df['energy_%'].hist(bins=30)
```

```python
# Obtener la correlación entre atributos como bpm, energy_%, valence_%, etc., y generar un heatmap de correlación.
df_filtrado = pysqldf("""
                    select * from df
                    where cast(streams as int) > 0
                      """)

corr = df_filtrado[['energy_%', 'valence_%', 'bpm', 'streams', 'acousticness_%', 'danceability_%']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
```
