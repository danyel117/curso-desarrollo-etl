```python
import pandas as pd

import pandasql as ps

import seaborn as sns
import matplotlib.pyplot as p
import matplotlib.pyplot as plt
import numpy as np
import scipy

def pysqldf(q):
    return ps.sqldf(q, globals())
```

```python
#Cargar el dataset usando pd.read_csv().

df = pd.read_csv('Spotify Most Streamed Songs.csv')
```

```python
# Mostrar las primeras 5 filas con df.head().

df.head()
```

```python
# Obtener información general del dataset, incluyendo tipos de datos, valores nulos, etc., usando df.info() y df.describe().
df.info()
```

```python
# Obtener información general del dataset, incluyendo tipos de datos, valores nulos, etc., usando df.info() y df.describe().
df.describe()
```

```python
# Identificar cuántos valores nulos hay por columna usando df.isnull().sum().


df.isnull().sum()
```

```python
pysqldf("""
        
        select 
        released_year,
        count(*)
        from
        df
        group by released_year
        """)
```

```python
# Distribución de canciones por año de lanzamiento:

df['released_year'].hist(bins=30)
```

```python
# Top 10 de artistas con más canciones:

top_10_artistas = pysqldf("""
             select 
             "artist(s)_name",
             count(*) as conteo              
             from df
             group by "artist(s)_name"
             order by conteo desc
             limit 10
             """)

top_10_artistas
```

```python
# Comparar el promedio de streams por modo musical:

promedio_por_modo = pysqldf("""
                            select 
                            mode,
                            avg(streams)
                            from df
                            group by mode
                            """)

promedio_por_modo
```

```python
# Crear un gráfico de dispersión para ver la relación entre la "danceability" y el número de streams.

danceability_vs_streams = pysqldf("""
                        select 
                        "danceability_%",
                        cast(streams as real) as streams
                        from df 
                        where cast(streams as real) > 0
                                  """)

sns.scatterplot(data=danceability_vs_streams, x='danceability_%', y='streams')
```

```python
# Agrupar por la columna key y calcular el promedio de streams para cada key.
pysqldf("""
        select
        key,
        avg(streams)
        from df
        group by key
        """)
```

```python
# Crear un histograma que muestre cómo se distribuye la energía de las canciones (energy_%).
df['energy_%'].hist(bins=30)
```

```python
filtered_df = pysqldf("""
                      select * from df
                      where cast(streams as real) > 0
                      """)

corr = filtered_df[['bpm', 'energy_%', 'valence_%', 'acousticness_%', 'danceability_%', 'streams']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
# Usar pandasql para seleccionar las 10 canciones con más streams.
pysqldf("""
        select
        *
        from
        df
        where cast(streams as real) > 0
        order by streams desc
        limit 10
        """)
```

```python
# Canciones lanzadas en el 2023 con más de 100 millones de streams:
pysqldf("""
        select * from 
        df
        where streams > 100000000
        and released_year = 2023
        """)

```

```python
# Calcular el bpm promedio de las canciones lanzadas entre 2020 y 2023

pysqldf("""
        select 
        released_year,
        avg(bpm)
        from
        df
        where released_year between 2020 and 2023
        group by released_year
        """)
```

```python
# Boxplot de la distribución de la "energy" por año de lanzamiento:
sns.boxplot(x='released_year', y='energy_%', data=df)
plt.xticks(rotation=45)
plt.xticks(ticks=range(0, len(df['released_year'].unique()), 5), rotation=45)

plt.show()

```

```python
# Gráfico de barras del promedio de "valence" por mes de lanzamiento:
sns.barplot(x='released_month', y='valence_%', data=df, estimator=np.mean)
plt.show()
```

```python
# Gráfico de calor (Heatmap) de la correlación entre atributos musicales:
corr = df[['bpm', 'danceability_%', 'energy_%', 'valence_%', 'acousticness_%']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Gráfico de violín de "speechiness" según el modo musical (mayor o menor)
sns.violinplot(x='mode', y='speechiness_%', data=df)
plt.xticks([0, 1], ['Minor', 'Major'])
plt.show()
```

```python
# FacetGrid de la relación entre "streams" y "bpm" por año de lanzamiento
facet_grid = pysqldf(
"""
select released_year, bpm, cast(streams as real) as streams from df where cast(streams as real) > 0
"""
)

g = sns.FacetGrid(facet_grid, col='released_year', col_wrap=4)
g.map(sns.scatterplot, 'bpm', 'streams')
plt.show()
```

```python
# Distribución de streams por released_year
filtered_df = pysqldf(
"""
select released_year, released_month, cast(streams as real) as streams from df where cast(streams as real) > 0
"""
)
sns.swarmplot(x='released_month', y='streams', data=filtered_df)
plt.xticks(rotation=45)
# plt.xticks(ticks=range(0, len(df['released_month'].unique()), 5), rotation=45)
plt.show()
```

```python
# Distribución de energy_% por released_year
sns.boxenplot(x='released_year', y='energy_%', data=df)
plt.xticks(rotation=45)
plt.xticks(ticks=range(0, len(df['released_year'].unique()), 5), rotation=45)
plt.show()
```

```python
# Cantidad de canciones lanzadas por released_month
sns.countplot(x='released_month', data=df)
plt.show()
```

```python
# Clustermap de la correlación entre atributos musicales
corr = df[['bpm', 'energy_%', 'valence_%', 'acousticness_%', 'danceability_%', 'speechiness_%']].corr()
sns.clustermap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Usando PandaSQL para filtrar los streams válidos
filtered_df = ps.sqldf(
"""
select 
    released_month, 
    cast(streams as real) as streams
from df 
where cast(streams as real) > 0
"""
)

# Crear la figura y los ejes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Gráfico de barras (countplot) en el eje izquierdo
sns.countplot(x='released_month', data=filtered_df, ax=ax1, color='skyblue')
ax1.set_ylabel('Cantidad de canciones', color='skyblue')
ax1.set_xlabel('Mes de lanzamiento')

# Crear un segundo eje para el gráfico de línea (plot de streams)
ax2 = ax1.twinx()

# Calcular el promedio de streams por mes, y asegurarse de rellenar meses sin datos
streams_per_month = filtered_df.groupby('released_month')['streams'].mean().reindex(range(1, 13), fill_value=0)

# Asegurar que los meses estén correctamente alineados
months = range(1, 13)  # Meses del 1 al 12

# Gráfico de línea en el segundo eje
ax2.plot(months, streams_per_month.values, color='red', marker='o')  # Usar los mismos meses
ax2.set_ylabel('Promedio de Streams', color='red')

plt.title('Cantidad de canciones lanzadas por mes y promedio de streams')
plt.show()
```
