```python
promedio_precios_por_modelo = pysqldf("""
        select
        Bike_Model,
        avg(Price) as promedio
        from df
        group by Bike_Model
        """)


df_sin_nulos = pysqldf("""
                       select 
                       df.*,
                       case
                        when df.Price is null then pm.promedio
                        else df.Price
                       end as PrecioModificado
                       from df     
                          join promedio_precios_por_modelo as pm
                                on df.Bike_Model = pm.Bike_Model
                       """)

df_sin_nulos.isnull().sum()
```

```python
# ¿Cuál es el top 3 de clientes que han generado más ingresos? ¿Cuánto ingreso generó cada uno de ellos?
total_venta_por_cliente = pysqldf("""
select
cl.nombre,
sum(co.venta) as total_venta        
from clientes as cl
    join contratos as co
        on co.clienteId = cl.id
group by cl.nombre
order by total_venta desc
-- limit 3
"""
)
```

```python
# ¿Cuál es el top 3 de usuarios con más horas reportadas en julio de 2024? ¿Cuántas horas reportó cada uno de ellos?
pysqldf("""
select
u.name,
u.lastName,
sum(r.horas) as total_horas        
from usuarios u
    join reportes r
        on r.usuarioId = u.id
where strftime('%Y', r.fecha) = '2024' and strftime('%m', r.fecha) = '07'
group by u.name, u.lastName
order by total_horas desc
limit 3
""")
```

```python
# ¿Cuáles son los 3 contratos con más horas reportadas durante 2024? ¿Cuántas horas se han reportado en cada uno de ellos?
pysqldf("""
select
c.nombre,
sum(r.horas) as total_horas        
from contratos c
    join reportes r
        on r.contratoId = c.id
where strftime('%Y', r.fecha) = '2024'
group by c.nombre
order by total_horas desc
limit 3
""")
```

```python
# ¿Cuál fue el mes con más horas trabajadas en 2023? ¿Cuántas horas se trabajaron en ese mes?
pysqldf("""
select 
strftime('%Y', fecha) as año, 
strftime('%m', fecha) as mes,    
sum(horas) as total_horas
from reportes
where strftime('%Y', fecha) = '2023'
group by
strftime('%Y', fecha), 
strftime('%m', fecha)
order by total_horas desc
limit 1
""")
```

```python
# ¿Cuáles son los dos usuarios con los reportes individuales más grandes y cuántas horas reportó cada uno?
pysqldf("""
        select
        u.name,
        u.lastName,
        max(r.horas) as horas_reportadas
        from usuarios u
            join reportes r
                on r.usuarioId = u.id
        group by u.name, u.lastName
        order by horas_reportadas desc
        limit 2
        """)
```

```python
# ¿Cuál es el top 3 de clientes con el mayor gasto de horas por contrato? ¿Cuál es el total de horas para cada uno de ellos?
pysqldf("""
select 
cl.nombre,
sum(r.horas) as total_horas    
from clientes cl
    join contratos co
        on co.clienteId = cl.id
    join reportes r
        on r.contratoId = co.id
group by cl.nombre
order by total_horas desc
limit 3
""")
```
